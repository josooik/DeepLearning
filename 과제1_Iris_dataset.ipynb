## 1. 데이터 정의
---

* sklearn 으로 부터 iris 데이터셋을 불러옵니다.
* x_data와 y_data를 정의합니다.

```python
import pandas as pd

data = pd.read_csv('iris_train.csv')

data
```

```python
import numpy as np

data_np = data.to_numpy()
print(data_np)
```

```python
x_data = data_np[:, 2:].astype(float)
y_data = data_np[:, 1]

print(x_data.shape, y_data.shape)

print("\n")

y_data
```

```python
y_data_label = []

for y in y_data:
  if y == 'setosa':
    y_data_label.append(0)

  elif y == 'versicolor':
    y_data_label.append(1)

  else:
    y_data_label.append(2)

print(y_data_label)
```

## 3. 데이터 전처리
---

* y 데이터 값을 원-핫 벡터로 코딩합니다.
* 학습데이터와 테스트데이터를 분리합니다. 
  * sklearn.model_selection에서 제공하는 train_test_split 를 사용합니다.


### **One-hot 코딩**


```python
from tensorflow.keras.utils import to_categorical

y_onehot = to_categorical(y_data_label, num_classes=3)

print(y_onehot)
```

### **train-test split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_onehot, test_size=0.2, random_state=42)

print(x_train.shape)
```

## 4. 모델학습
---

* softmax regression을 사용하여 훈련하고, 테스트셋을 이용해 성능을 평가합니다.
* decision tree를 사용하여 훈련하고, 테스트셋을 이용해 성능을 평가합니다.
* 이 데이터셋에 적합한 세번째 모델을 찾아보고 훈련해 봅니다.

### **Softmax regression**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

model = Sequential()
#model.add(Dense(4, input_dim=4, activation='relu'))
#model.add(Dense(3, activation='softmax'))

model.add(Dense(3, input_dim=4, activation='softmax'))
model.summary()

model.compile(optimizer = optimizers.SGD(learning_rate=0.01),
              loss = 'categorical_crossentropy',
              metrics=['acc'])

hist = model.fit(x_train, y_train, epochs=300)
```

```python
import matplotlib.pyplot as plt

# loss : 훈련 손실값
# acc : 훈련 정확도
# val_loss : 검증 손실값
# val_acc : 검증 정확도

plt.plot(hist.history['acc'])
#plt.plot(hist.history['val_acc'])
plt.title('Accuracy')
plt.legend(['train'], loc='upper left')
plt.show()

plt.plot(hist.history['loss'])
#plt.plot(hist.history['val_loss'])
plt.title('Loss')
plt.legend(['loss'], loc='upper left')
plt.show()
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
print("accuracy : ", score[1], "loss :", score[0])
```

### **Decision Tree**

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier()

dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

```python
tree.plot_tree(dt)
```
