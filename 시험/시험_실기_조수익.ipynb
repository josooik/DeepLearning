#### **데이터 전처리 (10)**
  * x_train, x_test 전처리: 정규화 (5)
  * y_train, y_test 전처리: 원핫벡터 (5)

```python
def minmax(x):
  x_min = np.min(x)
  x_max = np.max(x)
  print(x_min, x_max)

  return (x - x_min) / (x_max - x_min)
```

```python
x_train_minmax = minmax(x_train)
x_test_minmax = minmax(x_test)

print(x_train_minmax.shape, x_test_minmax.shape)
```

#### **cnn2 (5)** : 아래 그림의 모델을 구현합니다.
![](https://github.com/yebiny/Lecture_deeplearning_basic/blob/main/imgs/test_cnn2.png?raw=true)

```python
import tensorflow as tf
from tensorflow.keras import applications
from tensorflow.keras import layers, models
from tensorflow.keras import utils

model = models.Sequential()

def conv2d(x, filters=3):
  y = layers.Conv2D(filters, kernel_size=3, padding='same', activation='relu')(x)
  return y

def maxPooling2D(x, pool_size=(2, 2), stride=(4,4)):
  y = layers.MaxPooling2D(pool_size=pool_size, strides=stride)(x)
  return y

def flatten(x):
  y = layers.Flatten()(x)
  return y

def dense(x, dense=64):
  y = layers.Dense(dense)(x)
  return y

def concatenated(x):
  y = layers.Concatenate()(x)
  return y

x = layers.Input(shape=(32, 32, 3))
y = conv2d(x)
y = maxPooling2D(y)
y = conv2d(y)
y = maxPooling2D(y, stride=(2,2))
y = flatten(y)

d1 = dense(y)
d1 = dense(d1, dense=32)
d2 = dense(y)
d2 = dense(d2, dense=32)

print(d1.shape, d2.shape)

cont = concatenated([d1, d2])
y = dense(cont, dense=10)

model = models.Model(x, y)
model.summary()
utils.plot_model(model, show_shapes=True)
```

### **2-2. 모델 가져오기 (10)**

#### **모델 로드 후 시각화 (5)**
- tensorflow.keras의 applications에서 제공하는 모델 아무거나 로드합니다

```python
from tensorflow.keras import applications

base_model = applications.vgg16.VGG16(weights='imagenet',
                                     include_top=False,
                                     input_shape=[32, 32, 3])

base_model.summary()
utils.plot_model(base_model, show_shapes=True)
```

#### **가져온 모델 데이터에 맞게 수정후 시각화 (5)**

```python
from tensorflow.keras import layers, models

x = base_model.output
y = layers.Flatten(name='Add_flatten')(x)
y = layers.Dense(10, activation='softmax', name='Add_dense')(y)

model = models.Model(inputs=base_model.input, outputs=y)
model.summary()
utils.plot_model(model, show_shapes=True)
```
