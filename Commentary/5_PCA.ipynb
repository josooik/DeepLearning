## 공분산

## <center>$Var(\vec{x}) = \frac{1}{N} \sum_{i=0}^N(x_i - \bar{x})^2$</center>

## <center>$Cov({X}) = \frac{1}{N} X^{T}X$</center>


```python
def _var(x, ):
  N = len(x)
  x_mean = sum(x)/N
  return sum( [(xi-x_mean)**2 for xi in x] ) / len(x)
```

1. 데이터 정규화

```python
x = np.array([[-2,4],
              [-1,2],
              [0,0],
              [1,-1],
              [2,-5]])
```

2. 공분산행렬 구하기

```python
cov_x=np.cov(x.T)
print(cov_x)
cov_x = np.dot(x.T,x)/4
print(cov_x)
```

3. 공분산행렬의 eigen vlaue vector 구하기

```python
eig_vals, eig_vecs = np.linalg.eig(cov_x)
eig_vals, eig_vecs
```

4. 주성분 구하기 eign value 를 통해

```python
vec1 = eig_vecs[:,1]
```

5. 차원축소하기

```python
projected_x = np.dot(x, vec1)
projected_x
```
